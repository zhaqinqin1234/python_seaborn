```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.options.display.max_columns = 200
plt.style.use('ggplot')
```

```python
df = pd.read_excel("CombinedQ30.xlsx", sheet_name=2, header =0)
df.head()
```

```python
df.tail()
```

```python
total_df = df[df["level"]=="Total"]
```

```python
total_df.head()
```

```python
sns.boxplot(x="Flow Cell", y= "Yield Total (G)", data = df, hue="level")
```

```python
Reads_df = df[(df["level"]=="Read l") | (df["level"]=="Read 4")]
```

```python
sns.boxplot(x="Flow Cell", y="Yield Total (G)", data = Reads_df,hue="level")
```

# Total yield bar plot

```python
#set the color palette
palette = sns.color_palette("husl", len(total_df))

sns.barplot(x="SheetName", y="Yield Total (G)", data=total_df, palette=palette)
plt.title("Total yield between different runs")
plt.xlabel("Run Name")
plt.ylabel("Total yield (G)")
plt.xticks(rotation=45) # Rotate the x-axis tick label by 45 degree

#save the figure
plt.savefig("totalYield.png", bbox_inches="tight")
# Display the plot
plt.show()
```

# Total % Q30 plot

```python
#set the color palette
palette = sns.color_palette("bright", len(total_df))

sns.barplot(x="SheetName", y="% >= Q30", data=total_df, palette=palette)
plt.title("% >= Q30 between different runs")
plt.xlabel("Run Name")
plt.ylabel("% >= Q30")
plt.xticks(rotation=45) # Rotate the x-axis tick label by 45 degree

#save the figure
plt.savefig("Q30.png", bbox_inches="tight")
# Display the plot
plt.show()
```

```python
#set the color palette
palette = sns.color_palette("husl", len(total_df))

sns.barplot(x="SheetName", y="% >= Q30", data=total_df, palette=palette)
plt.title("% >= Q30 between different runs")
plt.xlabel("Run Name")
plt.ylabel("% >= Q30")
plt.xticks(rotation=45) # Rotate the x-axis tick label by 45 degree

# Add a horizontal line at y-80
plt.axhline(y=80, color ="r", linestyle='--', label="Threshold")
# Add a legned to the plot
plt.legend()

#save the figure
plt.savefig("Q30-with_hline.png", bbox_inches="tight")
# Display the plot
plt.show()
```
