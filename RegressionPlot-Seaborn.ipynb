```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import warnings
import numpy as np
warnings.filterwarnings("ignore")
```

```python
penguins = pd.read_csv("./sns_data/penguins.csv")
```

```python
sns.lmplot(  #使用figure-level方法lmplot()绘制回归关系图
    data=penguins,
    x="喙长 (毫米)",
    y="喙深 (毫米)",
    ci=None,
    line_kws={
        'linestyle': '-',
        'color': '#b1283a',
        'linewidth': 1,
        'alpha': 0.8
    },
    scatter_kws={
        'color': '#708ebf',
        's': 30,
        'alpha': 0.5
    },
)
```

## 10.13 回归关系图-离散数据+抖动
# 当x变量是离散的时，对散点图进行抖动处理。

```python
sns.lmplot(
    data=mpg,
    x="汽缸数",  #离散变量
    y="车身重量",
    x_jitter=.25,  #添加抖动
    line_kws={
        'linestyle': '-',
        'color': '#b1283a',
        'linewidth': 1,
        'alpha': 0.8
    },
    scatter_kws={
        'color': '#708ebf',
        's': 30,
        'alpha': 0.5
    },
    ci=95)
```

## 10.14 回归关系图-离散数据+聚合拟合

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 15))
functions = [np.mean, np.sum, np.max, np.min, np.median, np.std]
for i, function in enumerate(functions):
    ax = axs.flatten()[i]
    sns.regplot(
        data=mpg,
        x="汽缸数",
        y="加速度",
        line_kws={
            'linestyle': '-',
            'color': '#b1283a',
            'linewidth': 1,
            'alpha': 0.8
        },
        scatter_kws={
            'color': '#708ebf',
            's': 30,
            'alpha': 0.7
        },
        x_estimator=function,  #x_estimator设置聚合方式
        ax=ax)
    ax.set_title(function.__name__)

plt.tight_layout()  
plt.show() 
```
